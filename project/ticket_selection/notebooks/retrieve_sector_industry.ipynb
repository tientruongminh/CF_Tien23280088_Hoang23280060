```python
# Import dataset from labs/week123
import os
ticket_names = [name.split('.')[0] for name in os.listdir("/kaggle/input/computational-finance/yfinance/per_symbol")]
```

```python
from yahooquery import Ticker
import pandas as pd
from tqdm import tqdm

# Gọi yahooquery cho nhiều mã cùng lúc
t = Ticker(ticket_names)

# Lấy asset_profile: trả về dict lồng nhiều cấp
profiles = t.asset_profile  # key: ticker -> dict thông tin

rows = []
for symbol, info in tqdm(profiles.items()):
    if not isinstance(info, dict):
        continue
    rows.append({
        "ticker": symbol,
        "sectorKey": info.get("sectorKey"),
        "industryKey": info.get("industryKey"),
    })

df_sector = pd.DataFrame(rows)
# Lọc những mã không lấy được sector
df_sector = df_sector[~df_sector["sectorKey"].isna()]
print(df_sector.head())

df_sector.to_csv("sector_industry.csv")
```
