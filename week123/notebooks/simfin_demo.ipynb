# Cấu hình SimFin

```python
import simfin as sf
from simfin.names import *

# Đặt API key (lấy trong tài khoản SimFin)
sf.set_api_key('YOUR_API_KEY_HERE')

# Thư mục để SimFin lưu cache dữ liệu (tự động tạo nếu chưa tồn tại)
sf.set_data_dir('simfin_data')

# Thị trường muốn dùng (ở đây là US)
MARKET = 'us'

# Một ticker ví dụ để minh họa (Apple)
EXAMPLE_TICKER = 'AAPL'
```

# 2. Tải từng loại dataset

```python
# 2.1. Thông tin doanh nghiệp (companies)
companies = sf.load_companies(market=MARKET)
```

```python
# 2.2. Income Statement (Báo cáo kết quả kinh doanh)
# variant:
#   - 'annual'    báo cáo năm
#   - 'quarterly' báo cáo quý
#   - 'ttm'       trailing twelve months
income = sf.load_income(
    market=MARKET,
    variant='annual',
    index=[TICKER, REPORT_DATE],  # dùng multi index: (ticker, ngày báo cáo)
    parse_dates=True              # chuyển REPORT_DATE sang kiểu datetime
)
```

```python
# 2.3. Balance Sheet (Bảng cân đối kế toán)
balance = sf.load_balance(
    market=MARKET,
    variant='annual',
    index=[TICKER, REPORT_DATE],
    parse_dates=True
)
```

```python
# 2.4. Cashflow Statement (Báo cáo lưu chuyển tiền tệ)
cashflow = sf.load_cashflow(
    market=MARKET,
    variant='annual',
    index=[TICKER, REPORT_DATE],
    parse_dates=True
)
```

```python
# 2.5. Derived Metrics (các chỉ số tính toán sẵn: ROE, ROA, margin...)
try:
    # Thử TTM trước (nếu server / plan cho phép)
    derived = sf.load_derived(
        market=MARKET,
        variant='ttm',
        index=[TICKER, REPORT_DATE],
        parse_dates=True
    )
except Exception as e:
    print("Không load được derived TTM, chuyển sang annual. Lỗi chi tiết:")
    print(e)
    # Fallback: annual
    try:
        derived = sf.load_derived(
            market=MARKET,
            variant='annual',
            index=[TICKER, REPORT_DATE],
            parse_dates=True
        )
    except Exception as e2:
        print("Không load được derived annual luôn. Bỏ qua derived. Lỗi:")
        print(e2)
        derived = None
```

```python
# 2.6. Share Prices (giá cổ phiếu, OHLCV)
shareprices = sf.load_shareprices(
    market=MARKET,
    variant='daily',              # giá theo ngày
    index=[TICKER, 'Date'],       # index mặc định là (ticker, Date)
    parse_dates=True
)
```

# 3. In demo nhỏ gọn


```python
def show_basic_info():
    """In ra thông tin cơ bản của từng dataset."""

    print("\n=== COMPANIES (thông tin doanh nghiệp) ===")
    print("Shape:", companies.shape)
    print(companies.head())

    print("\n=== INCOME STATEMENT (annual) ===")
    print("Shape:", income.shape)
    print(income.head())

    print("\n=== BALANCE SHEET (annual) ===")
    print("Shape:", balance.shape)
    print(balance.head())

    print("\n=== CASHFLOW STATEMENT (annual) ===")
    print("Shape:", cashflow.shape)
    print(cashflow.head())

    print("\n=== DERIVED METRICS ===")
    if derived is not None:
        print("Shape:", derived.shape)
        print(derived.head())
    else:
        print("Không load được dataset derived.")


    print("\n=== SHARE PRICES (daily) ===")
    print("Shape:", shareprices.shape)
    print(shareprices.head())


def show_example_ticker(ticker: str = EXAMPLE_TICKER):
    """Lọc và in vài dòng cho một mã cụ thể, ví dụ AAPL."""

    print(f"\n\n######## EXAMPLE TICKER: {ticker} ########")

    # Income
    print("\n--- Income statement của mã", ticker, "(annual) ---")
    if (ticker,) in income.index:
        print(income.loc[(ticker,)].sort_index().tail())
    else:
        print("Không tìm thấy ticker trong income")

    # Balance
    print("\n--- Balance sheet của mã", ticker, "(annual) ---")
    if (ticker,) in balance.index:
        print(balance.loc[(ticker,)].sort_index().tail())
    else:
        print("Không tìm thấy ticker trong balance")

    # Cashflow
    print("\n--- Cashflow statement của mã", ticker, "(annual) ---")
    if (ticker,) in cashflow.index:
        print(cashflow.loc[(ticker,)].sort_index().tail())
    else:
        print("Không tìm thấy ticker trong cashflow")

    # Derived
    print("\n--- Derived metrics của mã", ticker, "---")
    if derived is not None and (ticker,) in derived.index:
        cols = [c for c in derived.columns if c.lower() in ['roe', 'roa', 'grossmargin', 'netmargin']]
        print(derived.loc[(ticker,)].sort_index()[cols].tail())
    else:
        print("Không có dữ liệu derived cho ticker này (hoặc không load được dataset).")


    # Share prices
    print("\n--- Share prices của mã", ticker, "(daily) ---")
    if (ticker,) in shareprices.index:
        print(shareprices.loc[(ticker,)].sort_index().tail())
    else:
        print("Không tìm thấy ticker trong shareprices")
```

```python
# In thông tin tổng quan của từng bảng
show_basic_info()
```

```python
# In ví dụ chi tiết cho một mã cụ thể
show_example_ticker(EXAMPLE_TICKER)
```
